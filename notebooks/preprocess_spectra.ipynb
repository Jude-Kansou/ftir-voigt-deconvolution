```python
import pandas as pd


# Load the Excel file
file_path = "zein_film_raw_data.xlsx"
data = pd.ExcelFile(file_path)

# Define the range of wavenumbers to keep
# wavenumber_min = 1480
# wavenumber_max = 1750

wavenumber_min = 1480
wavenumber_max = 1722

# List of sheet names to process
sheets = data.sheet_names

# Dictionary to store the filtered data from each sheet
filtered_data = {}

# Dictionary to store the filtered data from each sheet

# i = 0
for sheet in sheets:
    # print(i)
    # i+=1
    # Read the sheet and drop NaN values
    df = pd.read_excel(data, sheet).dropna()

    # Ensure column names are normalized (in case of hidden spaces or capitalization)
    df.columns = df.columns.str.strip().str.lower()

    # Check if "wavenumber" and "absorbance" columns exist
    if "wavenumber" in df.columns and "absorbance" in df.columns:
        # Filter data where wavenumber is between 1400 and 1800
        df_filtered = df[(df["wavenumber"] >= wavenumber_min) & (df["wavenumber"] <= wavenumber_max)]

        # # Store filtered data in a dictionary
        # if i > 3:
        #     print(df_filtered)

        df_filtered = pd.DataFrame(df_filtered)
        filtered_data[sheet] = df_filtered


output_file_path = "book1.xlsx"

# Create an Excel writer and save the filtered sheets
with pd.ExcelWriter(output_file_path) as writer:
    for sheet, df_filtered in filtered_data.items():

        df_filtered.to_excel(writer, sheet_name=sheet)
```
